```python
os.makedirs(OUTPUT_COLMAP, exist_ok=True)
os.makedirs(os.path.join(OUTPUT_COLMAP, 'images'), exist_ok=True)
video_extensiones = ['.mp4', '.avi', '.mov']
if os.path.isfile(INPUT_PATH) and Path(INPUT_PATH).suffix.lower() in video_extensiones:
    print("Procesando video...")
    cap = cv2.VideoCapture(INPUT_PATH)
    if not cap.isOpened():
        raise IOError(f"No se pudo abrir el video: {INPUT_PATH}")
    frame_count = 0
    image_idx = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if frame_count % 30 == 0:
            image_path = os.path.join(OUTPUT_COLMAP, 'images', f'frame_{image_idx:04d}.jpg')
            cv2.imwrite(image_path, frame)
            image_idx += 1
        frame_count += 1
    cap.release()
    print(f"Se extrajeron {image_idx} imágenes del video.")
elif os.path.isdir(INPUT_PATH):
    print("Usando directorio de imágenes...")
    for img_file in os.listdir(INPUT_PATH):
        full_file_name = os.path.join(INPUT_PATH, img_file)
        if os.path.isfile(full_file_name) and img_file.lower().endswith(('.png', '.jpg', '.jpeg')):
            shutil.copy(full_file_name, os.path.join(OUTPUT_COLMAP, 'images'))
else:
    raise ValueError("La ruta de entrada debe ser un video o un directorio de imágenes")
```

```python
print("Dividiendo el dataset y creando las carpetas y archivos JSON correspondientes...")
transforms_path = os.path.join(NERF_DATASET_DIR, 'transforms.json')
with open(transforms_path, 'r') as f:
    transforms = json.load(f)
frames = transforms['frames']
random.shuffle(frames)
num_frames = len(frames)
num_train = int(0.8 * num_frames)
num_val = int(0.1 * num_frames)
num_test = num_frames - num_train - num_val
train_frames = frames[:num_train]
val_frames = frames[num_train:num_train + num_val]
test_frames = frames[num_train + num_val:]
train_dir = os.path.join(NERF_DATASET_DIR, 'train')
val_dir = os.path.join(NERF_DATASET_DIR, 'val')
test_dir = os.path.join(NERF_DATASET_DIR, 'test')
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
def save_dataset_split(split_frames, split_dir, split_name):
    transforms_split = {
        'camera_angle_x': transforms['camera_angle_x'],
        'frames': []
    }
    for frame in split_frames:
        image_path = frame['file_path']
        image_name = os.path.basename(image_path)
        src_image = os.path.join(images_dir_for_nerf, image_name)
        dst_image = os.path.join(split_dir, image_name)
        shutil.copy(src_image, dst_image)
        frame['file_path'] = os.path.join('./', split_name, image_name)
        transforms_split['frames'].append(frame)
    transforms_split_path = os.path.join(NERF_DATASET_DIR, f'transforms_{split_name}.json')
    with open(transforms_split_path, 'w') as f:
        json.dump(transforms_split, f, indent=4)
save_dataset_split(train_frames, train_dir, 'train')
save_dataset_split(val_frames, val_dir, 'val')
save_dataset_split(test_frames, test_dir, 'test')
print(f"Dataset dividido y guardado en {NERF_DATASET_DIR}")
```

# ***MiDaS***

```python
images_dir = "/content/drive/MyDrive/SIS421/Examen Final - SIS421/Dataset" # @param {type:"string"}
```

```python
depth_maps_dir = "/content/drive/MyDrive/SIS421/Examen Final - SIS421/MiDaS/Profundidad" # @param {type:"string"}
```

```python
point_clouds_dir = "/content/drive/MyDrive/SIS421/Examen Final - SIS421/MiDaS/COLMAP" # @param {type:"string"}
```

```python
os.makedirs(depth_maps_dir, exist_ok=True)
os.makedirs(point_clouds_dir, exist_ok=True)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_type = "DPT_Large"
midas = torch.hub.load("intel-isl/MiDaS", model_type)
midas.to(device)
midas.eval()
midas_transforms = torch.hub.load("intel-isl/MiDaS", "transforms")
if model_type in ["DPT_Large", "DPT_Hybrid"]:
    transform = midas_transforms.dpt_transform
else:
    transform = midas_transforms.small_transform
```

```python
image_files = [f for f in os.listdir(images_dir) if f.endswith(('.png', '.jpg', '.jpeg'))]
for img_name in image_files:
    img_path = os.path.join(images_dir, img_name)
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    input_batch = transform(img_rgb).to(device)
    with torch.no_grad():
        prediction = midas(input_batch)
        prediction = torch.nn.functional.interpolate(
            prediction.unsqueeze(1),
            size=img_rgb.shape[:2],
            mode="bicubic",
            align_corners=False,
        ).squeeze()
    depth_map = prediction.cpu().numpy()
    depth_min = depth_map.min()
    depth_max = depth_map.max()
    depth_map_normalized = (depth_map - depth_min) / (depth_max - depth_min)
    depth_map_path = os.path.join(depth_maps_dir, img_name)
    plt.imsave(depth_map_path, depth_map_normalized, cmap='inferno')
```

```python
fx, fy = 1.0, 1.0  # Focal lengths
cx, cy = 0.0, 0.0  # Coordenadas del centro óptico
for img_name in image_files:
    img_path = os.path.join(images_dir, img_name)
    depth_map_path = os.path.join(depth_maps_dir, img_name)
    img = cv2.imread(img_path)
    depth_map = cv2.imread(depth_map_path, cv2.IMREAD_GRAYSCALE)
    h, w = depth_map.shape
    i_range = np.arange(h)
    j_range = np.arange(w)
    js, is_ = np.meshgrid(j_range, i_range)
    zs = depth_map / 255.0
    xs = (js - cx) * zs / fx
    ys = (is_ - cy) * zs / fy
    points = np.stack((xs, ys, zs), axis=-1).reshape(-1, 3)
    colors = img.reshape(-1, 3) / 255.0
    pcd = o3d.geometry.PointCloud()
    pcd.points = o3d.utility.Vector3dVector(points)
    pcd.colors = o3d.utility.Vector3dVector(colors)
    pcd_path = os.path.join(point_clouds_dir, img_name.replace('.jpg', '.ply'))
    o3d.io.write_point_cloud(pcd_path, pcd)
```

```python
pcd_combined = o3d.geometry.PointCloud()
for img_name in image_files:
    pcd_path = os.path.join(point_clouds_dir, img_name.replace('.jpg', '.ply'))
    pcd = o3d.io.read_point_cloud(pcd_path)
    pcd_combined += pcd
o3d.visualization.draw_geometries([pcd_combined])
```
